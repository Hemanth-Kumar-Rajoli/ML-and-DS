```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,8,10])
plt.ylabel("numbers")
plt.xlabel("indices")
plt.title("linear graph")
plt.show()
```

```python
plt.plot([1,2,3,4,5],[1,4,9,14,25])
plt.ylabel("squares")
plt.xlabel("numbers")
plt.title("square of a number")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4,5],[1,4,9,14,25],"ro")
plt.ylabel("squares")
plt.xlabel("numbers")
plt.title("square of a number")
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
points = np.arange(1,10,2)
points
```

```python
plt.plot(points,points*2,"g^",label="x*2")
plt.plot(points,points**2,"rs",label="x**2")
plt.plot(points,points*3,"b--",label='x*3')
plt.grid()
plt.legend()
plt.show()
```

```python
x1=[1,2,3,4]
y1=[2,4,6,8]
x2=[1,2,3,4]
y2=[4,8,12,16]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='green',linewidth=5)
plt.setp(lines[1],"linewidth",7,"color","red")
plt.show()
```

```python
t1 = np.array([0,30,45,90])
plt.figure()
plt.subplot(211)
plt.plot(t1,np.sin(t1),"r-")
plt.figure()
plt.subplot(212)
plt.plot(t1,np.cos(t1),"g--")
plt.show()
```
