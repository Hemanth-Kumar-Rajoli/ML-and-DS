```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = pd.read_csv('iris_dataset.csv')
iris_df.head()
```

```python
iris_df.columns
```

```python
iris_df.shape
```

```python
iris_df['target'].value_counts()
```

```python
iris_df.plot(kind="scatter",x='sepal length (cm)',y='sepal width (cm)')
plt.grid()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris_df,hue="target",height=4).map(plt.scatter,"sepal length (cm)","sepal width (cm)").add_legend()
plt.show()
```
