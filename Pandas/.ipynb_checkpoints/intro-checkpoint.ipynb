# Merge dataFrames - Like Joins in SQL

```python
temparature_dataFrame = pd.DataFrame({
    "Location":["mumbai","kadapa","delhi","bangalore"],
    "temparatures":[20,30,40,50]
})
temparature_dataFrame
```

```python
humidity_dataFrame=pd.DataFrame({
    "Location":["mumbai","kadapa","delhi","vizag"],
    "humidity":[1,2,4,5]
})
humidity_dataFrame
```

```python
pd.merge(temparature_dataFrame,humidity_dataFrame,on='Location')
```

```python
#outer join
pd.merge(temparature_dataFrame,humidity_dataFrame,on='Location',how='outer')
```
